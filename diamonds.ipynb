```python
X = df.drop(['price'], axis=1)
X.head()
```

```python
y = df['price']
y.head()
```

```python
columns = ['carat', 'cut', 'color', 'clarity', 'depth', 'table', 'x', 'y', 'z']
target = 'price'
```

```python
X= df[columns]
y= df[target]
```

```python
X.head(10)
```

```python
X_dummies = pd.get_dummies(X)
```

```python
X_dummies.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_dummies, y, random_state=42, test_size=0.33)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
sns.scatterplot(x=y_test , y=y_pred, color="red")
```

```python
sns.displot(y_pred - y_test, color="green")
```

```python
import matplotlib.pyplot as plt
plt.hist(y_pred[y_pred <0])
```

```python
print(lr.score(X_test, y_test))
```

```python
import joblib
```

```python
#saving the model
filename = 'diamond_regression_model.sav'
joblib.dump(lr, filename)
```

```python
#Loading the model
loaded_model = joblib.load(filename)
```
